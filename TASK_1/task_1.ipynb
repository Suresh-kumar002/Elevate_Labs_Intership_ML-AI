```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
# numerical
df['Fare'] = df['Fare'].fillna(df['Fare'].mode()[0])
 #categorical
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.distplot(df['Age'])
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Density")

plt.subplot(1, 2, 2)
sns.distplot(df['Fare'])
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```
